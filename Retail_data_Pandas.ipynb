```python
data["Age"]=data["Age"].fillna(data["Age"].median())
```

```python
data["Age"].unique()
```

```python
data=data[(data["Age"] >= 20) & (data["Age"] <= 65)]
```

```python
data["Order_Date_Year"]=data["Order_Date_Year"].fillna(data["Order_Date_Year"].mode()[0])
```

```python
data["Order_Date_Year"].isnull().sum()
```

```python
data["Order_Date_Year"]=data["Order_Date_Year"].astype(int)
```

```python
data.info()
```

```python
data["Gender"]=data["Gender"].str.strip().str.title()
```

```python
data["Region"]=data["Region"].str.strip().str.title()
```

```python
data["Region"].unique()
```

```python
data["Gender"].unique()
```

```python
data["Sales"]=data["Quantity"]*data["Unit_Price"] * (1 - data["Discount_%"] /100)
```

```python
data["Sales_Amount"] == data["Sales"]
```

```python
data["Sales_Amount"] = data["Sales"]
```

```python
data.drop("Sales",axis=1,inplace=True)
```

```python
data["Age"]
```

```python
def Age_Group(x):
    if x >= 20 and x <=30:
        return "Young"
    elif x >= 31 and x <=50:
        return "Adult"
    else:
        return "Senior"

data["Age_Group"]=data["Age"].apply(Age_Group)        
```

```python
data["Age_Group"]
```

```python
data["Sales_Amount"]
```

```python
def Revenue_Category(x):
    if x < 2000:
        return "Low"
    elif x > 2000 and x <= 10000:
        return "Medium"
    else:
        return "High"

data["Revenue_Category"]=data["Sales_Amount"].apply(Revenue_Category)
```

```python
data["Revenue_Category"]
```

```python
import numpy as np
```

```python
data["Log_Sales"]=np.log1p(data["Sales_Amount"])
```

```python
data["Log_Sales"]
```

```python
data[["Customer_ID","Age","Age_Group","Gender","Region","Order_Date_Year","Revenue_Category"]].to_csv("Customer_Info2.csv",index=False)
```

```python
data[["Order_ID","Customer_ID","Product_Category","Quantity","Unit_Price","Discount_%","Sales_Amount","Log_Sales","Payment_Mode","Delivery_Days","Customer_Rating","Returned"]].to_csv("Sales_Data.csv",index=False)
```
