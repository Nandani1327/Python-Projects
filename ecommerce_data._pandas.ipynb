```python
data["Signup_Year"]=data["Signup_Year"].astype(int)
```

```python
data["total"]=data["Quantity"]*data["Price_Per_Unit"]*(1-data["Discount_Percent"]/100)
```

```python
data["Total_Amount"]==data["total"]
```

```python
data["Total_Amount"]=data["total"]
```

```python
data.drop("total",axis=1,inplace=True)
```

```python
def Age_Group(x):
    if x >=18 and x <= 25:
        return "Young"
    elif x >=26 and x <= 45:
        return "Adult"
    else:
        return "Senior"

data["Age_Group"]=data["Age"].apply(Age_Group)
```

```python
data["Age_Group"]
```

```python
data["Total_Amount"]
```

```python
def Revenue_Category(x):
    if x < 1000:
        return "Low"
    elif x >= 1000 and x <= 4000:
        return "Medium"
    else:
        return "High"

data["Revenue_Category"]=data["Total_Amount"].apply(Revenue_Category)
```

```python
import numpy as np
```

```python
data["Log_Total_Amount"]=np.log1p(data["Total_Amount"])
```
